```python
import pandas as pd
import requests
from io import StringIO

# URLs for the JHU CSSE COVID-19 data (example for global data)
confirmed_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_confirmed_global.csv"
deaths_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_deaths_global.csv"
recovered_url = "https://raw.githubusercontent.com/CSSEGISandData/COVID-19/master/csse_covid_19_data/csse_covid_19_time_series/time_series_covid19_recovered_global.csv"

# Function to load data
def load_jhu_data(url):
    response = requests.get(url)
    response.raise_for_status() # Raise an exception for bad status codes
    df = pd.read_csv(StringIO(response.text))
    return df

print("Downloading COVID-19 data...")
confirmed_df = load_jhu_data(confirmed_url)
deaths_df = load_jhu_data(deaths_url)
recovered_df = load_jhu_data(recovered_url)
print("Download complete.")

print("\nConfirmed Cases Data Head:")
display(confirmed_df.head())

print("\nDeaths Data Head:")
display(deaths_df.head())

print("\nRecovered Cases Data Head:")
display(recovered_df.head())
```

```python
# Identify date columns (all columns after 'Long')
def get_date_columns(df):
    return df.columns[df.columns.get_loc('Long') + 1:]

confirmed_dates = get_date_columns(confirmed_df)
deaths_dates = get_date_columns(deaths_df)
recovered_dates = get_date_columns(recovered_df)

# Melt the dataframes to long format
id_vars = ['Province/State', 'Country/Region', 'Lat', 'Long']

confirmed_melted = confirmed_df.melt(id_vars=id_vars, value_vars=confirmed_dates, var_name='Date', value_name='Confirmed')
deaths_melted = deaths_df.melt(id_vars=id_vars, value_vars=deaths_dates, var_name='Date', value_name='Deaths')
recovered_melted = recovered_df.melt(id_vars=id_vars, value_vars=recovered_dates, var_name='Date', value_name='Recovered')

# Merge the melted dataframes
# First, merge confirmed and deaths
full_df = pd.merge(confirmed_melted, deaths_melted, on=id_vars + ['Date'], how='left')
# Then, merge with recovered
full_df = pd.merge(full_df, recovered_melted, on=id_vars + ['Date'], how='left')

# Convert 'Date' column to datetime objects
full_df['Date'] = pd.to_datetime(full_df['Date'])

# Fill any NaN values in Recovered with 0 (as JHU stopped reporting recovered for many regions)
full_df['Recovered'] = full_df['Recovered'].fillna(0)

print("Merged and preprocessed data head:")
display(full_df.head())

print("\nInformation about the merged dataframe:")
full_df.info()
```

```python
# Aggregate by Country/Region and Date
country_daily_df = full_df.groupby(['Country/Region', 'Date']).agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum'
}).reset_index()

# Sort by Country/Region and Date for correct 'daily new' calculations
country_daily_df = country_daily_df.sort_values(by=['Country/Region', 'Date'])

# Calculate daily new cases, new deaths, and new recoveries
country_daily_df['New_Confirmed'] = country_daily_df.groupby('Country/Region')['Confirmed'].diff().fillna(0)
country_daily_df['New_Deaths'] = country_daily_df.groupby('Country/Region')['Deaths'].diff().fillna(0)
country_daily_df['New_Recovered'] = country_daily_df.groupby('Country/Region')['Recovered'].diff().fillna(0)

# Ensure new cases are not negative (can happen due to data corrections/reclassification)
for col in ['New_Confirmed', 'New_Deaths', 'New_Recovered']:
    country_daily_df[col] = country_daily_df[col].apply(lambda x: max(0, x))

# Calculate Active Cases
country_daily_df['Active_Cases'] = country_daily_df['Confirmed'] - country_daily_df['Deaths'] - country_daily_df['Recovered']
country_daily_df['Active_Cases'] = country_daily_df['Active_Cases'].apply(lambda x: max(0, x)) # Ensure no negative active cases

# Display the head of the aggregated data
print("Aggregated data with daily new cases and active cases:")
display(country_daily_df.head())

print("\nInformation about the aggregated dataframe:")
country_daily_df.info()
```

```python
# Aggregate globally by Date
global_daily_df = country_daily_df.groupby('Date').agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum',
    'New_Confirmed': 'sum',
    'New_Deaths': 'sum',
    'New_Recovered': 'sum',
    'Active_Cases': 'sum' # Sum active cases globally
}).reset_index()

# Display the head of the global daily data
print("Global daily summary data head:")
display(global_daily_df.head())

print("\nInformation about the global daily dataframe:")
global_daily_df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set_style('whitegrid')
plt.figure(figsize=(15, 10))

# Plot Global Daily New Confirmed Cases
plt.subplot(3, 1, 1)
sns.lineplot(x='Date', y='New_Confirmed', data=global_daily_df, color='blue')
plt.title('Global Daily New Confirmed Cases')
plt.xlabel('')
plt.ylabel('New Cases')
plt.tight_layout()

# Plot Global Daily New Deaths
plt.subplot(3, 1, 2)
sns.lineplot(x='Date', y='New_Deaths', data=global_daily_df, color='red')
plt.title('Global Daily New Deaths')
plt.xlabel('')
plt.ylabel('New Deaths')
plt.tight_layout()

# Plot Global Active Cases
plt.subplot(3, 1, 3)
sns.lineplot(x='Date', y='Active_Cases', data=global_daily_df, color='orange')
plt.title('Global Active Cases')
plt.xlabel('Date')
plt.ylabel('Active Cases')
plt.tight_layout()

plt.show()
```

```python
# Calculate Recovery Rate and Death Rate
# Avoid division by zero by checking if Confirmed is 0
global_daily_df['Recovery_Rate'] = (global_daily_df['Recovered'] / global_daily_df['Confirmed'] * 100).fillna(0)
global_daily_df['Death_Rate'] = (global_daily_df['Deaths'] / global_daily_df['Confirmed'] * 100).fillna(0)

# Ensure rates are not NaN if Confirmed is 0 for that day
global_daily_df.loc[global_daily_df['Confirmed'] == 0, 'Recovery_Rate'] = 0
global_daily_df.loc[global_daily_df['Confirmed'] == 0, 'Death_Rate'] = 0

plt.figure(figsize=(15, 6))

# Plot Global Recovery Rate
plt.subplot(2, 1, 1)
sns.lineplot(x='Date', y='Recovery_Rate', data=global_daily_df, color='green')
plt.title('Global Recovery Rate Over Time')
plt.xlabel('')
plt.ylabel('Recovery Rate (%)')
plt.tight_layout()

# Plot Global Death Rate
plt.subplot(2, 1, 2)
sns.lineplot(x='Date', y='Death_Rate', data=global_daily_df, color='red')
plt.title('Global Death Rate Over Time')
plt.xlabel('Date')
plt.ylabel('Death Rate (%)')
plt.tight_layout()

plt.show()

print("Global daily data with calculated rates head:")
display(global_daily_df.head())
```

```python
# Get the last date in the dataset
last_date = country_daily_df['Date'].max()

# Filter for the latest data for each country
country_latest_df = country_daily_df[country_daily_df['Date'] == last_date].copy()

# Select relevant columns and reset index if needed
country_latest_df = country_latest_df[['Country/Region', 'Confirmed', 'Deaths', 'Recovered', 'Active_Cases']].reset_index(drop=True)

# Calculate latest recovery and death rates for each country
country_latest_df['Recovery_Rate'] = (country_latest_df['Recovered'] / country_latest_df['Confirmed'] * 100).fillna(0)
country_latest_df['Death_Rate'] = (country_latest_df['Deaths'] / country_latest_df['Confirmed'] * 100).fillna(0)

# Ensure rates are 0 if Confirmed is 0 for that country
country_latest_df.loc[country_latest_df['Confirmed'] == 0, 'Recovery_Rate'] = 0
country_latest_df.loc[country_latest_df['Confirmed'] == 0, 'Death_Rate'] = 0

print("Latest cumulative data per country head:")
display(country_latest_df.head())

print("\nInformation about the country latest dataframe:")
country_latest_df.info()
```

```python
# Top 10 Countries by Confirmed Cases
top_confirmed = country_latest_df.sort_values('Confirmed', ascending=False).head(10)
print("\nTop 10 Countries by Confirmed Cases:")
display(top_confirmed[['Country/Region', 'Confirmed']])

# Top 10 Countries by Deaths
top_deaths = country_latest_df.sort_values('Deaths', ascending=False).head(10)
print("\nTop 10 Countries by Deaths:")
display(top_deaths[['Country/Region', 'Deaths']])

# Top 10 Countries by Active Cases
top_active = country_latest_df.sort_values('Active_Cases', ascending=False).head(10)
print("\nTop 10 Countries by Active Cases:")
display(top_active[['Country/Region', 'Active_Cases']])

# Visualization for Top Confirmed Cases
plt.figure(figsize=(12, 6))
sns.barplot(x='Country/Region', y='Confirmed', data=top_confirmed, palette='viridis')
plt.title('Top 10 Countries by Confirmed Cases')
plt.xlabel('Country/Region')
plt.ylabel('Total Confirmed Cases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualization for Top Deaths
plt.figure(figsize=(12, 6))
sns.barplot(x='Country/Region', y='Deaths', data=top_deaths, palette='mako')
plt.title('Top 10 Countries by Deaths')
plt.xlabel('Country/Region')
plt.ylabel('Total Deaths')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualization for Top Active Cases
plt.figure(figsize=(12, 6))
sns.barplot(x='Country/Region', y='Active_Cases', data=top_active, palette='rocket')
plt.title('Top 10 Countries by Active Cases')
plt.xlabel('Country/Region')
plt.ylabel('Total Active Cases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Top 10 Countries by Recovery Rate (excluding countries with 0 confirmed cases to avoid misleading rates)
top_recovery = country_latest_df[country_latest_df['Confirmed'] > 0].sort_values('Recovery_Rate', ascending=False).head(10)
print("\nTop 10 Countries by Recovery Rate:")
display(top_recovery[['Country/Region', 'Recovery_Rate', 'Confirmed']])

# Bottom 10 Countries by Recovery Rate (excluding countries with 0 confirmed cases)
bottom_recovery = country_latest_df[country_latest_df['Confirmed'] > 0].sort_values('Recovery_Rate', ascending=True).head(10)
print("\nBottom 10 Countries by Recovery Rate:")
display(bottom_recovery[['Country/Region', 'Recovery_Rate', 'Confirmed']])

# Top 10 Countries by Death Rate (excluding countries with 0 confirmed cases)
top_death_rate = country_latest_df[country_latest_df['Confirmed'] > 0].sort_values('Death_Rate', ascending=False).head(10)
print("\nTop 10 Countries by Death Rate:")
display(top_death_rate[['Country/Region', 'Death_Rate', 'Confirmed']])


# Visualization for Top Recovery Rate
plt.figure(figsize=(12, 6))
sns.barplot(x='Country/Region', y='Recovery_Rate', data=top_recovery, palette='Greens_r')
plt.title('Top 10 Countries by Recovery Rate')
plt.xlabel('Country/Region')
plt.ylabel('Recovery Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()

# Visualization for Top Death Rate
plt.figure(figsize=(12, 6))
sns.barplot(x='Country/Region', y='Death_Rate', data=top_death_rate, palette='Reds_r')
plt.title('Top 10 Countries by Death Rate')
plt.xlabel('Country/Region')
plt.ylabel('Death Rate (%)')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```
